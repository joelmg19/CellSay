```python
# ==============================================
# üß© 1Ô∏è‚É£ INSTALACI√ìN DE DEPENDENCIAS
# ==============================================
!pip install roboflow ultralytics

# ==============================================
# üîë 2Ô∏è‚É£ CONECTAR CON GOOGLE DRIVE
# ==============================================
from google.colab import drive
drive.mount('/content/drive')

# Carpeta donde guardaremos los resultados
PROJECT_DIR = "/content/drive/MyDrive/YOLO_Carteles"
!mkdir -p {PROJECT_DIR}

# ==============================================
# üì¶ 3Ô∏è‚É£ DESCARGAR DATASET DESDE ROBOFLOW
# ==============================================
from roboflow import Roboflow

API_KEY = key  # <-- Pega aqu√≠ tu clave Roboflow
WORKSPACE = "mtrharry"
PROJECT_NAME = "deteccion-de-carteles"
VERSION = 2

rf = Roboflow(api_key=API_KEY)
project = rf.workspace(WORKSPACE).project(PROJECT_NAME)
version = project.version(VERSION)
dataset = version.download("yolov8")

print("‚úÖ Dataset descargado en:", dataset.location)
```

```python

# ==============================================
# ü§ñ 4Ô∏è‚É£ ENTRENAR EL MODELO YOLOv8
# ==============================================
from ultralytics import YOLO

# Cargar modelo base
model = YOLO("yolov8n.pt")

# Entrenar el modelo (guarda en Google Drive)
results = model.train(
    data=dataset.location + "/data.yaml",
    epochs=100,
    imgsz=640,
    batch=16,
    project=PROJECT_DIR,
    name="entrenamiento_carteles"
)

print("‚úÖ Entrenamiento finalizado. Resultados guardados en:", PROJECT_DIR)
```

```python
# ==============================================
# üß™ 5Ô∏è‚É£ VALIDAR EL MODELO
# ==============================================
model.val()
```

```python
from google.colab import files
from IPython.display import Image, display

# Subir una imagen desde tu PC
uploaded = files.upload()

# Tomar el nombre del archivo
for filename in uploaded.keys():
    image_path = filename

# Hacer la predicci√≥n
results = model.predict(source=image_path, show=True)

# Mostrar resultado
display(Image(filename=image_path))
print("‚úÖ Predicci√≥n completada:", results)
```
